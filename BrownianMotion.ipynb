```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
random = [0]
for i in range(100):
    random.append(rnd.uniform(-1,1))
plt.ylabel('Range')    
plt.xlabel('Number of Steps')
plt.plot(random)
plt.show()
```

```python
Brownian = [0]
for i in range(1000):
    Brownian.append(Brownian[i]+rnd.uniform(-1,1))
    
plt.ylabel('Range')    
plt.xlabel('Number of Steps')
plt.plot(Brownian)
plt.show()
```

```python
run = []

for j in range(100000):
    Brownian = [0]
    dist = 0
    for i in range(100):
        step = rnd.uniform(-1,1)
        Brownian.append( Brownian[i] + step )
        if step > 0:
            dist = dist + 1
        elif step < 0:
            dist = dist - 1
        
    run.append(dist)
    #print(dist)

    

plt.ylabel('Range')    
plt.xlabel('Brownian')
plt.plot(Brownian)
plt.show()
```

```python
run.count(0)
```

```python
plt.hist(run , bins = 60)
plt.title('Histogram Brownian Motion Displacements ')
plt.show()
```

```python
def f(x):
    return x**2 - 2

orbits = [0.1]    #Start at 0.1
for i in range(20000):
    orbits.append(f(orbits[i]))
    
plt.plot(orbits)
plt.show()
```

```python
plt.hist(orbits , bins = 400)
plt.title('Histogram of the orbit of 0.01 under $x^2-2$ ')
plt.show()
```

```python
import random
def rollDice(n):
    """ Rolls n dice and returns the sum """
    total = 0
    for i in range(n):
        total += random . randint (1, 6)
    return total
```

```python
rolls = []
numRolls = 100000
for i in range(numRolls):
    rolls.append(rollDice(6))
```

```python
for i in range(6,36+1):
    print(i," has ",int(rolls.count(i)))
```
