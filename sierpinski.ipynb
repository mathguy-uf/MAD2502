```python
#NAME: (Please type name here)

#%matplotlib notebook

import matplotlib.pyplot as plt
import random

#initalize lists to starting point (0,0)
L_x = [0]
L_y = [0]

def next_pt(x,y,num):
    if num == 0:
        L_x.append(x/2.0)
        L_y.append(y/2.0)
    elif num == 1:
        L_x.append((x+1)/2.0)
        L_y.append(y/2.0)
    else:
        L_x.append((x+0.5)/2)
        L_y.append((y+ (3**0.5) ) / 2.0 )
        

for t in range(20):
    n = random.randint(0, 2) #Randomly picks 0, 1, 2
    next_pt( L_x[-1], L_y[-1], n) #computes next point
    
#Use matplotlib to plot L_x and L_y where L_x has the x-coordinates and L_y has the y-coorinates. 
#Print them up as small disjoint points.  

plt.plot(L_x, L_y, 'bo', markersize=10)
plt.show()
```
