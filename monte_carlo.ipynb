```python
import random as rnd
import matplotlib.pyplot as plt

max=100000
in_circ=0   #counter for points in circle
in_sqr=max  #all points are in the square

#list of points in the circle
Cx=[]
Cy=[]
#list of points in the square but not in the circle
Sx=[]
Sy=[]

for i in range(max):
    x=rnd.random()   #picks a random x value in [0,1]
    y=rnd.random()   #picks a random y value in [0,1]
    
    if x**2+y**2<1:  #if the point is in the circle
        in_circ+=1   #adds 1 to number of points in the circle
        Cx.append(x) 
        Cy.append(y)
    else:
        Sx.append(x)
        Sy.append(y)
        

        
        
#makes aspect Ration 1.0
plt.figure(figsize=(10, 10))

#plot both scatter plots together
plt.scatter(Cx,Cy,c='red',label='Inside the Circle')
plt.scatter(Sx,Sy,c='blue',label='Outside the Circle')

#label the figure
plt.title('Monte Carlo Method for Computing Pi')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

#place the legend box
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 0.5))

#plot everything together
plt.show()

#approximation of pi
approx_pi=4*in_circ/in_sqr

print(max,"number of random points were placed in the square and",in_circ,"were inside the circle. \n")
print("Pi is approximately:",approx_pi)
```
